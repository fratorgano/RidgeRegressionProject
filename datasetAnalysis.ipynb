# Data Analysis

### track_id

```python
print(f"Each value is unique? {df['track_id'].size == df['track_id'].unique().size}")
print(df['track_id'].shape[0])
print(df['track_id'].nunique())
show(df['track_id'].value_counts())
grouped = df.groupby('track_id')
for (id,group) in grouped:
  if group['popularity'].shape[0]>1:
    if(not group['popularity'].eq(group['popularity'].iloc[0]).all()):
      mean = group['popularity'].mean()
      if(mean>(group['popularity'].iloc[0]+1) or mean<(group['popularity'].iloc[0]-1)):
        print(id,group['popularity'].tolist(), mean)
```
